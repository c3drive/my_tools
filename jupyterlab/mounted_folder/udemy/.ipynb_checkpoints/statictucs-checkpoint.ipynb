```python
import seaborn as sns
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
# seabornは13でFigure-level API が自動描画しないことが増えた。
# その時は最後に.figをつけるか、plt.show()をする
sns.__version__
```

## データ準備

```python
df = sns.load_dataset('tips')
df['tip_rate'] = df['tip'] / df['total_bill']
df
```

## 分布の描画

```python
# seabornは13でFigure-level API が自動描画しないことが増えた。
# その時は最後に.figをつけるか、plt.show()をする
sns.displot(df['tip'], kde=False).fig
```

```python
sns.catplot(x=df['time'], data=df, kind='count').fig
```

## 平均値

```python
np.mean(df['tip_rate'])
```

```python
df['tip_rate'].mean()
```

```python
# df.groupby(df['sex'], observed=False).mean(numeric_only=True)['tip_rate']でもOK
df.groupby(df['sex'], observed=False)['tip_rate'].mean()
```

```python
sns.barplot(x='sex', y='tip_rate', data=df, errorbar=None)
plt.show()
```

```python
df.groupby(df['sex'], observed=False)['tip_rate'].mean().plot(kind='bar')
plt.show()
```

## 中央値

```python
np.median(df['tip_rate'])
```

```python
df['tip_rate'].median()
```

```python
df.groupby('sex', observed=False).median(numeric_only=True)
```

```python
sns.barplot(x='sex', y='tip_rate', data=df, estimator=np.median, errorbar=None)
plt.show()
```

```python
df.sort_values('tip_rate', ascending=False)
```

## 最頻値

```python
mode, count = stats.mode(df['tip_rate'])
print(mode)
print(count)
```

```python
# 最頻値が2つある場合も出力できる
df['tip_rate'].mode()
```

## 分散と標準偏差
データがどれくらいばらついているか
- 平均から±標準偏差は、データの約2/3が入っていると言われている
- 平均から±2標準偏差は、データの約95%が入っていると言われている
- 平均から±3標準偏差は、データの役99-100%が入っていると言われている

```python
# 分散
np.var(df['tip_rate'])
```

```python
# 標準偏差
np.std(df['tip'])
```

```python
np.sqrt(np.var(df['tip']))
```

```python
# 平均どれくらいだったか
tip_mean = df['tip'].mean()
tip_mean
```
